```python
#q9 -  Create a random vector of size 30 and find the mean value
arr = np.random.random(30)

x, y = 0, 0

while x*y != 30:
    x = random.randint(1, 31)
    y = random.randint(1, 31)

arr = arr.reshape(x, y)

print(arr)

print("\nMean of the Array is " + str(arr.mean()))
```

```python
#q10 - Create a numpy array having elements 0 to 10 And negate all the elements between 3 and 9
arr= np.arange(0,10)
print(arr)
arr = [(-d if (d > 3 and d < 9) else d) for d in arr]
print(arr)
```

```python
#q14 - Create a random matrix and Compute a matrix rank
np_array = np.random.randint(0,9,16).reshape(2, 2, 2, 2)
print(np_array)

array = np.array([4, 2, 7, 1])
temp = array.argsort()
ranks = np.empty_like(temp)
ranks[temp] = np.arange(len(array))

print(ranks)
```
